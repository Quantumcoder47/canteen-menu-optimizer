## Import Libraries & Load Dataset
We start by importing Python libraries and loading the dataset.

```python
# === Environment & Imports ===
import sys, os  # Importing required libraries
import numpy as np  # Importing required libraries
import pandas as pd  # Importing required libraries
import matplotlib.pyplot as plt  # Importing required libraries
import seaborn as sns  # Importing required libraries
```

```python
# sklearn
import sklearn  # Importing required libraries
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_score  # Importing required libraries
from sklearn.pipeline import Pipeline  # Importing required libraries
from sklearn.compose import ColumnTransformer  # Importing required libraries
from sklearn.impute import SimpleImputer  # Importing required libraries
from sklearn.preprocessing import StandardScaler, OneHotEncoder  # Importing required libraries
from sklearn.linear_model import LogisticRegression  # Importing required libraries
from sklearn.ensemble import RandomForestClassifier  # Importing required libraries
from sklearn.metrics import (classification_report, confusion_matrix,  # Importing required libraries
                             ConfusionMatrixDisplay, accuracy_score, f1_score)  # Calculating accuracy of the model
from sklearn.inspection import permutation_importance  # Importing required libraries
```

```python
# imbalanced-learn (only if used)
try:
    import imblearn  # Importing required libraries
    from imblearn.pipeline import Pipeline as ImbPipeline  # Importing required libraries
    from imblearn.over_sampling import SMOTE, RandomOverSampler  # Importing required libraries
except Exception:
    imblearn = None

import joblib  # Importing required libraries
from sklearn.exceptions import ConvergenceWarning  # Importing required libraries
import warnings  # Importing required libraries

RANDOM_STATE = 42
warnings.filterwarnings("ignore", category=ConvergenceWarning)

print("Python", sys.version.splitlines()[0])
print("pandas", pd.__version__, "sklearn", sklearn.__version__)
```

```python
# === Load & clean dataset ===
df = pd.read_csv("canteen_data.csv")  # Reading dataset from CSV file
def clean_col(c):
    return (str(c).strip()
            .replace("\n"," ")  # Mapping or replacing categorical values
            .replace("  "," ")  # Mapping or replacing categorical values
            .lower()
            .replace(" ", "_")  # Mapping or replacing categorical values
            .replace("(", "")  # Mapping or replacing categorical values
            .replace(")", "")  # Mapping or replacing categorical values
            .replace("/", "_"))  # Mapping or replacing categorical values
df.columns = [clean_col(c) for c in df.columns]
```

## Preprocessing

**What this section does:**
- Calculates BMI using the formula: $$\mathrm{BMI} = \frac{\text{weight (kg)}}{\text{height (m)}^2}$$
- Encodes categorical variables (LabelEncoder / One-Hot as used in the original notebook).
- Scales numerical features (StandardScaler / MinMaxScaler as used).

All transformations preserve original code logic; comments explain each step.

```python
# Age fix
if "age" in df.columns:
    df["age"] = df["age"].astype(str).str.extract(r"(\d+)")[0]  # Converting datatype of column
    df["age"] = pd.to_numeric(df["age"], errors="coerce").astype("Int64")  # Converting datatype of column
```

```python
# Ensure numeric height/weight
for col in ("height","weight"):
    if col in df.columns:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(r"[^\d\.]", "", regex=True), errors="coerce")  # Converting datatype of column
```

```python
# BMI
if "height" in df.columns and "weight" in df.columns:
    df["bmi"] = df["weight"] / ((df["height"]/100)**2)
    df["bmi"] = df["bmi"].round(2)
```

```python
# Cuisine count
for c in ("cuisine_top1","cuisine_top2","cuisine_top3"):
    if c in df.columns:
        df[c] = df[c].astype(str).replace("nan", pd.NA).replace("", pd.NA).str.strip()  # Converting datatype of column
cuisine_cols = [c for c in ("cuisine_top1","cuisine_top2","cuisine_top3") if c in df.columns]
if cuisine_cols:
    df["num_cuisines_recorded"] = df[cuisine_cols].notna().sum(axis=1)
```

```python
# Numeric coercion
for c in ("spice_tolerance","sweet_tooth_level","eating_out_per_week","food_budget_per_meal_inr"):
    if c in df.columns:
        df[c] = pd.to_numeric(df[c].astype(str).str.replace(r"[^\d\.]", "", regex=True), errors="coerce")  # Converting datatype of column
```

```python
# Target clean
if "dietary_preference" in df.columns:
    df["dietary_preference"] = df["dietary_preference"].astype(str).str.strip().str.title()  # Converting datatype of column
    df = df[df["dietary_preference"].notna() & (df["dietary_preference"] != "Nan")]
```

```python
# Drop exact duplicates and save
df = df.drop_duplicates().reset_index(drop=True)
df.to_csv("canteen_data_clean.csv", index=False)
print("Clean shape:", df.shape)
df.head()  # Displaying first few rows of dataset
```

```python
# === Preprocessing pipeline ===
numerical_features = [c for c in ["age","bmi","spice_tolerance","sweet_tooth_level",
                                  "eating_out_per_week","food_budget_per_meal_inr",
                                  "num_cuisines_recorded","spice_sweet_interaction"] if c in df.columns]
categorical_features = [c for c in ["cuisine_top1"] if c in df.columns]
```

```python
# create interaction if available
if "spice_tolerance" in df.columns and "sweet_tooth_level" in df.columns and "spice_sweet_interaction" not in df.columns:
    df["spice_sweet_interaction"] = df["spice_tolerance"] * df["sweet_tooth_level"]

numeric_transformer = Pipeline(steps=[  # Creating ML pipeline (preprocessing + model)
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
categorical_transformer = Pipeline(steps=[  # Creating ML pipeline (preprocessing + model)
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, numerical_features),
    ("cat", categorical_transformer, categorical_features)
], remainder="drop")

print("Numerical features:", numerical_features)
print("Categorical features:", categorical_features)
```

## Train/Test Split

```python
# === Train/test split ===
X = df[numerical_features + categorical_features]
y = df["dietary_preference"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE)  # Splitting dataset into train and test sets
print("Train/test shapes:", X_train.shape, X_test.shape)
```

```python
# === Baseline: Logistic Regression ===
pipe_lr = Pipeline(steps=[  # Creating ML pipeline (preprocessing + model)
    ("preprocessor", preprocessor),
    ("clf", LogisticRegression(solver="lbfgs", max_iter=5000, class_weight="balanced", random_state=RANDOM_STATE))  # Initializing Logistic Regression model
])
pipe_lr.fit(X_train, y_train)  # Training the model
y_pred_lr = pipe_lr.predict(X_test)  # Making predictions on test data
print("LR Accuracy:", accuracy_score(y_test, y_pred_lr))  # Calculating accuracy of the model
print("LR Macro F1:", f1_score(y_test, y_pred_lr, average="macro"))
print(classification_report(y_test, y_pred_lr, zero_division=0))  # Generating classification report
```

```python
# === RandomForest + GridSearchCV ===
pipe_rf = Pipeline(steps=[  # Creating ML pipeline (preprocessing + model)
    ("preprocessor", preprocessor),
    ("clf", RandomForestClassifier(random_state=RANDOM_STATE, class_weight="balanced"))  # Initializing Random Forest Classifier model
])

param_grid = {
    "clf__n_estimators": [100, 300],
    "clf__max_depth": [None, 10, 20],
    "clf__min_samples_leaf": [1, 2]
}
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=RANDOM_STATE)
grid_rf = GridSearchCV(pipe_rf, param_grid, cv=cv, scoring="f1_macro", n_jobs=-1, verbose=1)  # Performing hyperparameter tuning with GridSearchCV
grid_rf.fit(X_train, y_train)  # Training the model
best_rf = grid_rf.best_estimator_
print("Best params:", grid_rf.best_params_)
```

```python

# === Evaluate best RandomForest on test set ===
est = best_rf
y_pred = est.predict(X_test)  # Making predictions on test data
print("RF Accuracy:", accuracy_score(y_test, y_pred))  # Calculating accuracy of the model
print("RF Macro F1:", f1_score(y_test, y_pred, average="macro"))
print(classification_report(y_test, y_pred, zero_division=0))  # Generating classification report

cm = confusion_matrix(y_test, y_pred, labels=est.classes_)  # Generating confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=est.classes_)  # Generating confusion matrix
disp.plot(cmap="Blues", xticks_rotation=45)
plt.title("Confusion Matrix (Test Set)")
plt.savefig("plots/confusion_matrix.png", dpi=150)
plt.show()
```

## Feature Importance & Permutation Importance

```python
# === Feature importances & permutation importance (robust) ===
est = best_rf

all_feature_names = est.named_steps["preprocessor"].get_feature_names_out()
n_features = int(est.named_steps["clf"].n_features_in_)
feature_names = all_feature_names[:n_features]

print("Total preprocessor features:", len(all_feature_names))
print("Features used by model:", n_features)
```

```python
rf = est.named_steps["clf"]
imp = pd.Series(rf.feature_importances_, index=feature_names).sort_values(ascending=False)  # Getting feature importance from Random Forest
print("\nTop RF importances:")
print(imp.head(15))  # Displaying first few rows of dataset
```

```python
imp.head(15).plot(kind="barh", figsize=(8,6)); plt.gca().invert_yaxis(); plt.title("RF Feature Importances"); 
plt.savefig("plots/random_forest_feature_importance.png", dpi=150)
plt.show()  # Displaying first few rows of dataset
```

```python
perm = permutation_importance(est, X_test, y_test, n_repeats=20, random_state=RANDOM_STATE, n_jobs=-1)  # Calculating permutation importance for explainability
perm_values = perm.importances_mean
perm_feature_names = feature_names[:len(perm_values)]
perm_imp = pd.Series(perm_values, index=perm_feature_names).sort_values(ascending=False)
print("\nTop Permutation importances:")
print(perm_imp.head(15))  # Displaying first few rows of dataset
```

```python

perm_imp.head(15).plot(kind="barh", figsize=(8,6)); plt.gca().invert_yaxis(); plt.title("Permutation Importances"); 
plt.savefig("plots/permutation_importance.png", dpi=150)
plt.show()  # Displaying first few rows of dataset
```
